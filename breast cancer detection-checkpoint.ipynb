Let's import the necessary libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.utils import shuffle
```

Let's have a look at the data

```python
df = pd.read_csv(r'D:\joel\UDEMY -\datasets and codes\Notebooks\.ipynb_checkpoints\breast cancer detection\data.csv')
print (df.shape)
df.head(8)
```

The columns 'id' &'Unnamed: 32' are just index & Null value variables. Let's remove them before we go any further.

```python
df = df.drop(['id','Unnamed: 32'],axis=1)
df.shape
```

We can see the dataset has many variables but very few observations.

Let's have a look at the distribution of Benign:0 Vs Malignant:1 observations

```python
print(df['diagnosis'].value_counts()/df.shape[0]*100)
df['diagnosis'].value_counts().plot(kind='bar')
```

We can see that more than 67% of the observations are Benign & 37% are Malignant.

This is an imbalanced dataset, we can try to balance these observations after Outlier removal & Feature selection.

Let's also replace the encodings 'B' with 0 and 'M' with 1

```python
df['diagnosis'] = df['diagnosis'].replace(['M'],1)
df['diagnosis'] = df['diagnosis'].replace(['B'],0)
print((df['diagnosis'].value_counts()/df.shape[0])*100)
```

Let's remove the Outlier then.

### Outlier Removal

```python
for i in df.columns[1:]:
    q25,q75 = np.percentile(df[i],[25,75])
    iqr = q75 - q25
    maximum = q75 + (iqr*1.5)
    minimum = q25 - (iqr*1.5)
    df.loc[df[i]>maximum,i]= np.NaN
    df.loc[df[i]<minimum,i]= np.NaN
```

Check for missing values after Outlier-removal

```python
(df.isnull().sum()/df.shape[0])*100
```

There are many missing values after Outlier removal, let's impute them

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
df_array = np.array(df)
df = pd.DataFrame(data = imputer.fit_transform(df_array),columns = df.columns)
df.isnull().sum()
```

Let's look at the distribution of the independent continuous variables

### Data Visualization

```python
for col in df.columns[1:]:
    sns.distplot(df[col],norm_hist=True,color='blue',kde_kws={'label':'KDE'}) # histogram with the kernel-distribution line.
    plt.axvline(df[col].mean(), color = 'r', linestyle = 'dashed', linewidth = 2) # plotting mean line
    plt.axvline(df[col].median(),color='b',linewidth=2) # plotting median line
    plt.show()
```

In almost every variable, the mean & median are close, even coinciding in some. The Outlier removal must have helped so.

Let's perform Feature selection techniques over the independent variables w.r.t target 'diagnosis'

### ANOVA test

let's perform ANOVA (analysis of Variance) test between the categorical-target and the independent numeric variables

```python
related=[]
unrelated =[]#  list of variables  not-related to the target

from scipy import stats
for i in df.columns[1:]:
    fStat, pVal = stats.f_oneway(df['diagnosis'], df[i])
    #print(i,' P-value: ',pVal,' f-statistic: ',fStat)
    if pVal<0.05:
        related.append(i)
    else:
        unrelated.append(i)    
        
print ('related:',related) 
print ('unrelated:',unrelated)              
```

ANOVA test says only 'radius_se' variable is not related to the target 'diagnosis'

### ExtraTrees Classifier

it uses a tree-based classifiers approach

```python
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.ensemble import ExtraTreesClassifier

# on the red wine dataset
X = df.iloc[:,1:]
y = df['diagnosis']
Importances = ExtraTreesClassifier().fit(X,y)
#print (redImportances.feature_importances_)

# Let's look at the scores visually

print ('Extra trees- feature importance')
Values = pd.Series(Importances.feature_importances_, index = df.columns[1:])
Values.nlargest(df.shape[1]-1).plot(kind='bar')
plt.figsize=(10,15)
plt.show()
```

Extra trees classifier too shows importances of different variables in varying proportions.

### Logistic regression feature importance

```python
# let's perform both tests on red wine first
# source: https://towardsdatascience.com/feature-selection-using-regularisation-a3678b71e499
from sklearn.linear_model import Lasso, LogisticRegression # Lasso is L1 regularization
from sklearn.feature_selection import SelectFromModel

# fitting the feature selection model with the Logistic regression data
sel_ = SelectFromModel(LogisticRegression(C=1, penalty='l2'))
sel_.fit(X,y)


# Visualising features that were kept by the lasso regularisation
print (sel_.get_support())

#  So True is for the features that lasso thought is important (non-zero features) 
# while False is for the features whose weights were shrinked to zero and are not important according to Lasso.


selected_feat = X.columns[(sel_.get_support())] # gives the remaining features
LR_unimportant=[i for i in df.columns[1:] if i not in selected_feat]

print ('important: ',selected_feat)
print ('unimportant: ',LR_unimportant)
```

Let's normalize the continuous variables between 0 and 1

### Feature Normalization

```python
for c in df2.columns[1:]:
    df2[c] = (df2[c]-min(df2[c]))/(max(df2[c])-min(df2[c]))
    print (c,': ',min(df2[c]),max(df2[c]))
    
df2.head() 
```

Let's split the data into train & test.

```python
from sklearn.model_selection import train_test_split
from collections import Counter

X = df2.iloc[:,1:]
y = df2['diagnosis']

# Let's split the data into train & test before oversampling the training data
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7)
print ('Training shape: ',X_train.shape,y_train.shape)
print ('Testing shape: ',X_test.shape,y_test.shape)
print ('Ratio of 1 & 0 in ytest:',Counter(y_test))
```

As we saw earlier, the count of malignant 1 observations is very low compared to benign cases 0, we need to balance the dataset.
Also we need to oversample the training data after balancing so that model has enough observations to train properly.

```python
from imblearn.over_sampling import RandomOverSampler
os=RandomOverSampler(1)
X_train_os,y_train_os = os.fit_sample(X_train,y_train)
print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes after fit {}".format(Counter(y_train_os)))
print ('New training data shape: ',X_train_os.shape,y_train_os.shape)
```

Let's oversample the data 4 times and look at it's shape

```python
X0 = np.array(X_train_os[y_train_os==0])
X1 = np.array(X_train_os[y_train_os==1])
X = np.vstack([X0, X0, X0, X0, X1, X1, X1, X1])
Y = np.concatenate(([0]*4*len(X0), [1]*4*len(X1)))
X,Y = shuffle(X,Y)
X.shape,Y.shape
```

Let's define an error-metric function to judge the model's classifications

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

def error_metric(y_true,y_preds):
    CM = pd.crosstab(y_true,y_preds) # the confusion matrix
    TN = CM.iloc[0,0]
    FN = CM.iloc[1,0]
    TP = CM.iloc[1,1]
    FP = CM.iloc[0,1]
    print (classification_report(y_true, y_preds))
    print (confusion_matrix(y_true,y_preds))
    print ('accuracy score: ',accuracy_score(y_true, y_preds)*100,'%') #check accuracy of model
    print ('False-Negative-Rate: ',(FN*100)/(FN+TP),'%') #False Negative rate 
    print ('true-positive-rate: ',((TP)*100)/(TP+FN),'%')
```

Let's also define a function to plot the AUC-ROC curve

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score
from matplotlib import pyplot as plt
def plotRocAuc(fpr, tpr, thresholds,score):
       
    # Plot the "dumb model" line
    plt.plot([0, 1], [0, 1], linestyle='--')
    
    # Plot the model line
    plt.plot(fpr, tpr, marker='.')
    plt.text(0.75, 0.25, "AUC: " + str(round(score,2)))
    
    # show the plot
    plt.show()
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR_model = LogisticRegression().fit(X,Y) 
LR_preds = LR_model.predict_proba(X_test)
print (error_metric(y_test,np.round((LR_preds)[:,1])))
LR_fpr,LR_tpr,LR_thresholds = roc_curve(y_test,LR_preds[:,1])
LR_score = roc_auc_score(y_test, LR_preds[:,1])
plotRocAuc(LR_fpr, LR_tpr, LR_thresholds,LR_score)
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
DT_model = DecisionTreeClassifier(criterion='entropy').fit(X,Y)
DT_preds = DT_model.predict_proba(X_test)
print(error_metric(y_test,np.round(DT_preds[:,1])))
DT_fpr, DT_tpr, DT_thresholds = roc_curve(y_test, DT_preds[:,1])
DT_score = roc_auc_score(y_test, DT_preds[:,1])
plotRocAuc(DT_fpr, DT_tpr, DT_thresholds,DT_score)
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
RF_model = RandomForestClassifier(n_estimators=100).fit(X,Y)
RF_preds = RF_model.predict_proba(X_test)
error_metric(y_test,np.round(RF_preds[:,1]))
RF_fpr,RF_tpr,RF_thresholds = roc_curve(y_test,RF_preds[:,1])
RF_score = roc_auc_score(y_test, RF_preds[:,1])
plotRocAuc(RF_fpr, RF_tpr, RF_thresholds, RF_score)
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
NB_model = GaussianNB().fit(X,Y)
NB_preds = NB_model.predict_proba(X_test)
print (error_metric(y_test,np.round((NB_preds)[:,1])))
NB_fpr,NB_tpr,NB_thresholds = roc_curve(y_test,NB_preds[:,1])
NB_score = roc_auc_score(y_test, NB_preds[:,1])
plotRocAuc(NB_fpr, NB_tpr, NB_thresholds, NB_score)
```

### K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
K_model = KNeighborsClassifier(n_neighbors=5).fit(X,Y)
K_preds = K_model.predict_proba(X_test)
print (error_metric(y_test,np.round((K_preds)[:,1])))
K_fpr,K_tpr,K_thresholds = roc_curve(y_test,K_preds[:,1])
K_score = roc_auc_score(y_test, K_preds[:,1])
plotRocAuc(K_fpr, K_tpr, K_thresholds, K_score)
```

### AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
ada_classifier=AdaBoostClassifier().fit(X,Y)
ADA_preds = ada_classifier.predict_proba(X_test)
ADA_fpr, ADA_tpr, ADA_thresholds = roc_curve(y_test, ADA_preds[:,1])
ADA_score = roc_auc_score(y_test, ADA_preds[:,1])
error_metric(y_test,np.round(ADA_preds[:,1]))
plotRocAuc(ADA_fpr, ADA_tpr, ADA_thresholds, ADA_score)
```

### Support Vector Classifier

```python
from sklearn.svm import SVC
SV_model = SVC(C=1.0,probability=True).fit(X,Y)
SV_preds = SV_model.predict_proba(X_test)

SV_fpr, SV_tpr, SV_thresholds = roc_curve(y_test, SV_preds[:,1])
SV_score = roc_auc_score(y_test, SV_preds[:,1])
error_metric(y_test,np.round(SV_preds[:,1]))
plotRocAuc(SV_fpr, SV_tpr, SV_thresholds, SV_score)
```

### Neural Network

```python
from keras import models
from keras import layers

def build_model():
    model = models.Sequential()
    model.add(layers.Dense(64, activation='sigmoid', input_shape=(X.shape[1],)))
    model.add(layers.Dense(64, activation='sigmoid'))
    model.add(layers.Dense(1, activation='sigmoid')) # because of multi-class labels
    # defining the optimizer and loss and accuracy parameters
    model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
    return model

# testing how the individual model with sampLe parameters scores on test data
X_train_os2, x_val, y_train_os2, y_val = train_test_split(X,Y,train_size=0.7)
print (X_train_os2.shape, y_train_os2.shape)
print (x_val.shape,y_val.shape)
model = build_model()
history = model.fit(X_train_os2, y_train_os2, batch_size=64, epochs=5, validation_data=(x_val, y_val))

DL_preds = model.predict_proba(X_test)
print (DL_preds.shape)
test_results = model.evaluate(X_test,y_test)
for i in range(len(test_results)):
    print (model.metrics_names[i],': ',test_results[i])

DL_preds2 = [round(i[0]) for i in DL_preds]


import matplotlib.pyplot as plt
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

DL_fpr, DL_tpr, DL_thresholds = roc_curve(y_test, DL_preds)
DL_score = roc_auc_score(y_test, DL_preds)
error_metric(y_test,np.array(DL_preds2))
plotRocAuc(DL_fpr, DL_tpr, DL_thresholds, DL_score)
```
